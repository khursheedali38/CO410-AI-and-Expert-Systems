```python
import cv2
import os
import random
from matplotlib import pyplot as plt
from glob import glob
import pandas as pd
import numpy as np
```

```python
# load the data
PATH = os.path.abspath(os.path.join('.', 'Dataset'))
MELANOMA_IMAGES = os.path.join(PATH, 'Melanoma')
MELANOMA_AUG = os.path.join(PATH, 'Melanoma_aug')
OTHER_IMAGES = os.path.join(PATH, 'others')
GT = os.path.join(PATH, 'gt')
images_others = glob(os.path.join(OTHER_IMAGES, '*.jpg'))
images_melanoma = glob(os.path.join(MELANOMA_IMAGES, '*.jpg'))
images_gt = glob(os.path.join(GT, '*.png'))
images_mel_aug = glob(os.path.join(MELANOMA_AUG, '*.jpg'))
```

```python
# resizing the images into size 128 x 128
def proc_images():
    x = []
    WIDTH = 128
    HEIGHT = 128
    count = 1
    
    for img in images_others:
        print("Processing Others image " + str(count))
        full_size_image = cv2.imread(img)
        x.append(cv2.resize(full_size_image, (WIDTH, HEIGHT), interpolation=cv2.INTER_CUBIC))
        count+=1
    
    for img in images_melanoma:
        print("Processing Melanoma image " + str(count))
        full_size_image = cv2.imread(img)
        x.append(cv2.resize(full_size_image, (WIDTH, HEIGHT), interpolation=cv2.INTER_CUBIC))
        count+=1

    for img in images_mel_aug:
        print("Processing Melanoma Augmented image " + str(count))
        full_size_image = cv2.imread(img)
        x.append(cv2.resize(full_size_image, (WIDTH, HEIGHT), interpolation=cv2.INTER_CUBIC))
        count+=1
    
    return x
```

```python
x = proc_images()
```

# Setting up in Model Keras

```python
from keras.preprocessing.image import ImageDataGenerator
from keras import applications
from keras import layers, models, optimizers
from keras import backend as K
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense
```

```python
K.image_data_format()
```

```python
# use pretrained vgg model
img_rows, img_cols, img_channel = 128, 128, 3
base_model = applications.vgg16.VGG16(include_top=False, weights='imagenet', input_tensor=None, input_shape=(img_rows, img_cols, img_channel), pooling=None)
```

```python
img_width, img_height = 128, 128
nb_train_samples = len(X_train)
nb_validation_samples = len(X_val)
epochs = 50
batch_size = 16
```

```python
model = models.Sequential()

model.add(layers.Flatten(input_shape=base_model.output_shape[1:]))
model.add(layers.Dense(512))
model.add(layers.BatchNormalization())
model.add(layers.Activation("relu"))

model.add(layers.Dropout(0.2))

model.add(layers.Dense(1))
model.add(layers.BatchNormalization())
model.add(layers.Activation("sigmoid"))

model = models.Model(inputs=base_model.input, outputs=model(base_model.output))
model.compile(
    loss='binary_crossentropy',
    optimizer='adadelta',
    metrics=['acc'])

model.summary()
```

# Data Augmentation
##### Rescaling the values by 1/255
##### Randomly flipping images horizontally
##### Randomly rotating in a range of 30 degrees

```python
train_datagen = ImageDataGenerator(rescale=1. / 255, horizontal_flip=True, rotation_range=30)
valtest_datagen = ImageDataGenerator(rescale=1. / 255)
train_generator = train_datagen.flow(np.array(X_train), y_train, batch_size=batch_size)
validation_generator = valtest_datagen.flow(np.array(X_val), y_val, batch_size=batch_size)
test_generator = valtest_datagen.flow(np.array(X_test), y_test, batch_size=batch_size)
```

```python
history = model.fit_generator(
    train_generator, 
    steps_per_epoch=nb_train_samples // batch_size,
    epochs=epochs,
    validation_data=validation_generator,
    validation_steps=nb_validation_samples // batch_size
)

model.save_weights('weights.h5')
```

```python
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(acc) + 1)
plt.plot(epochs, acc, 'blue', label='Training acc')
plt.plot(epochs, val_acc, 'red', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()
plt.figure()
plt.plot(epochs, loss, 'blue', label='Training loss')
plt.plot(epochs, val_loss, 'red', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()
plt.show()
```

```python
# confusion matrix, normalized
import sklearn
import itertools
from sklearn.metrics import confusion_matrix
dict_characters = {0: 'Benign', 1: 'Malignant'}
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    plt.figure(figsize = (5,5))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
a=X_train
b=y_train
c=X_valtest
d=y_valtest
Y_pred = model.predict(c)
Y_pred_classes = np.argmax(Y_pred,axis=1) 
cnf = confusion_matrix(d, Y_pred_classes) 
plot_confusion_matrix(cnf, classes = list(dict_characters.values()))
plt.show()

```
